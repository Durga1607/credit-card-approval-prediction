```python
# ============================================================
# CREDIT CARD APPROVAL PREDICTION SYSTEM
# Step 1 : Import Libraries & Load Dataset
# ============================================================

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os

warnings.filterwarnings("ignore")

# Display Settings
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 1000)

# ------------------------------------------------------------
# Project Path
# ------------------------------------------------------------

project_path = r"C:\Users\pmonu\OneDrive\DPK\project\smart bridge"

dataset_path = os.path.join(project_path, "Dataset")

application_path = os.path.join(dataset_path, "application_record.csv")
credit_path = os.path.join(dataset_path, "credit_record.csv")

# ------------------------------------------------------------
# Load Dataset
# ------------------------------------------------------------

application_df = pd.read_csv(application_path)
credit_df = pd.read_csv(credit_path)

# ------------------------------------------------------------
# Basic Information
# ------------------------------------------------------------

print("="*70)
print("      CREDIT CARD APPROVAL PREDICTION SYSTEM")
print("="*70)

print("\nApplication Dataset Shape :", application_df.shape)
print("Credit Dataset Shape      :", credit_df.shape)

print("\nApplication Dataset")
display(application_df.head())

print("\nCredit Dataset")
display(credit_df.head())

print("\nApplication Dataset Information")
application_df.info()

print("\nCredit Dataset Information")
credit_df.info()

print("\nMissing Values in Application Dataset")
print(application_df.isnull().sum())

print("\nMissing Values in Credit Dataset")
print(credit_df.isnull().sum())

print("\nDuplicate Rows")
print("Application Dataset :", application_df.duplicated().sum())
print("Credit Dataset      :", credit_df.duplicated().sum())

print("\nUnique Values in Application Dataset")
for column in application_df.columns:
    print(f"{column:<25} : {application_df[column].nunique()}")

print("\nCredit Status Distribution")
print(credit_df["STATUS"].value_counts())

print("\nStatistical Summary of Application Dataset")
display(application_df.describe(include="all"))

print("\nStatistical Summary of Credit Dataset")
display(credit_df.describe(include="all"))

print("\nDataset Loaded Successfully.")
```

```python
# ============================================================
# Step 2 : Data Cleaning
# ============================================================

print("="*70)
print("DATA CLEANING")
print("="*70)

# ------------------------------------------------------------
# 1. Check Missing Values
# ------------------------------------------------------------

print("\nMissing Values Before Cleaning")
print("-"*70)

missing_values = application_df.isnull().sum()
missing_values = missing_values[missing_values > 0]

if len(missing_values) == 0:
    print("No Missing Values Found")
else:
    print(missing_values)

# ------------------------------------------------------------
# 2. Fill Missing Values
# ------------------------------------------------------------

print("\nHandling Missing Values...")

# Fill Occupation Type with 'Unknown'
application_df["OCCUPATION_TYPE"] = application_df["OCCUPATION_TYPE"].fillna("Unknown")

print("Missing Values Handled Successfully.")

# ------------------------------------------------------------
# 3. Check Duplicate Records
# ------------------------------------------------------------

print("\nChecking Duplicate Records")

app_duplicates = application_df.duplicated().sum()
credit_duplicates = credit_df.duplicated().sum()

print(f"Application Dataset Duplicate Rows : {app_duplicates}")
print(f"Credit Dataset Duplicate Rows      : {credit_duplicates}")

# ------------------------------------------------------------
# 4. Remove Duplicate Records
# ------------------------------------------------------------

application_df.drop_duplicates(inplace=True)
credit_df.drop_duplicates(inplace=True)

print("\nDuplicate Records Removed.")

# ------------------------------------------------------------
# 5. Dataset Shape After Cleaning
# ------------------------------------------------------------

print("\nDataset Shape After Cleaning")

print("Application Dataset :", application_df.shape)
print("Credit Dataset      :", credit_df.shape)

# ------------------------------------------------------------
# 6. Final Missing Values Check
# ------------------------------------------------------------

print("\nRemaining Missing Values")

print(application_df.isnull().sum())

# ------------------------------------------------------------
# 7. Data Types
# ------------------------------------------------------------

print("\nApplication Dataset Data Types")

display(application_df.dtypes)

print("\nCredit Dataset Data Types")

display(credit_df.dtypes)

print("\nData Cleaning Completed Successfully.")
```

```python
# ============================================================
# Step 3 : Create Target Variable
# ============================================================

print("="*70)
print("CREATING TARGET VARIABLE")
print("="*70)

# ------------------------------------------------------------
# Bad Payment Status
# ------------------------------------------------------------

bad_status = ['1', '2', '3', '4', '5']

# ------------------------------------------------------------
# Create Target for Every Customer
# ------------------------------------------------------------

target_df = (
    credit_df
    .groupby("ID")["STATUS"]
    .apply(lambda x: 0 if any(status in bad_status for status in x) else 1)
    .reset_index()
)

# Rename Target Column

target_df.rename(columns={"STATUS": "TARGET"}, inplace=True)

# ------------------------------------------------------------
# Display Sample
# ------------------------------------------------------------

print("\nTarget Dataset")

display(target_df.head())

# ------------------------------------------------------------
# Target Distribution
# ------------------------------------------------------------

print("\nTarget Distribution")

print(target_df["TARGET"].value_counts())

# ------------------------------------------------------------
# Percentage
# ------------------------------------------------------------

print("\nTarget Percentage")

print(round(target_df["TARGET"].value_counts(normalize=True)*100,2))

print("\nTarget Variable Created Successfully.")
```

```python
# ============================================================
# Step 4 : Merge Application Dataset with Target Dataset
# ============================================================

print("="*70)
print("MERGING DATASETS")
print("="*70)

# ------------------------------------------------------------
# Merge Dataset
# ------------------------------------------------------------

final_df = application_df.merge(target_df, on="ID", how="inner")

# ------------------------------------------------------------
# Display Dataset Information
# ------------------------------------------------------------

print("\nFinal Dataset Shape")
print(final_df.shape)

print("\nFirst Five Records")
display(final_df.head())

print("\nLast Five Records")
display(final_df.tail())

# ------------------------------------------------------------
# Check Missing Values
# ------------------------------------------------------------

print("\nMissing Values")

display(final_df.isnull().sum())

# ------------------------------------------------------------
# Check Duplicate Records
# ------------------------------------------------------------

print("\nDuplicate Rows :", final_df.duplicated().sum())

# ------------------------------------------------------------
# Dataset Information
# ------------------------------------------------------------

print("\nDataset Information")

final_df.info()

print("\nTarget Distribution")

display(final_df["TARGET"].value_counts())

print("\nTarget Percentage")

display(round(final_df["TARGET"].value_counts(normalize=True)*100,2))

print("\nMerge Completed Successfully.")
```

```python
# ============================================================
# Step 5 : Final Dataset Verification
# ============================================================

print("="*80)
print("FINAL DATASET VERIFICATION")
print("="*80)

# Shape
print("\nDataset Shape")
print(final_df.shape)

# Columns
print("\nColumn Names")
print(final_df.columns.tolist())

# Data Types
print("\nData Types")
display(final_df.dtypes)

# Missing Values
print("\nMissing Values")
display(final_df.isnull().sum())

# Duplicate Records
print("\nDuplicate Rows :", final_df.duplicated().sum())

# Target Distribution
print("\nTarget Distribution")
display(final_df["TARGET"].value_counts())

print("\nTarget Percentage")
display(round(final_df["TARGET"].value_counts(normalize=True) * 100, 2))

print("\nDataset Verification Completed Successfully.")
```

```python
# ============================================================
# Step 6 : Exploratory Data Analysis (EDA)
# ============================================================

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

print("="*80)
print("EXPLORATORY DATA ANALYSIS")
print("="*80)

# ------------------------------------------------------------
# Figure Size
# ------------------------------------------------------------
plt.rcParams["figure.figsize"] = (8,5)

# ------------------------------------------------------------
# 1. Target Distribution
# ------------------------------------------------------------

plt.figure()
sns.countplot(x="TARGET", data=final_df)
plt.title("Target Distribution")
plt.xlabel("Approval Status (1=Approved, 0=Rejected)")
plt.ylabel("Count")
plt.show()

# ------------------------------------------------------------
# 2. Gender Distribution
# ------------------------------------------------------------

plt.figure()
sns.countplot(x="CODE_GENDER", data=final_df)
plt.title("Gender Distribution")
plt.show()

# ------------------------------------------------------------
# 3. Income Type
# ------------------------------------------------------------

plt.figure(figsize=(10,5))
sns.countplot(y="NAME_INCOME_TYPE", data=final_df)
plt.title("Income Type Distribution")
plt.show()

# ------------------------------------------------------------
# 4. Education Level
# ------------------------------------------------------------

plt.figure(figsize=(10,5))
sns.countplot(y="NAME_EDUCATION_TYPE", data=final_df)
plt.title("Education Level")
plt.show()

# ------------------------------------------------------------
# 5. Housing Type
# ------------------------------------------------------------

plt.figure(figsize=(10,5))
sns.countplot(y="NAME_HOUSING_TYPE", data=final_df)
plt.title("Housing Type")
plt.show()

# ------------------------------------------------------------
# 6. Occupation Type
# ------------------------------------------------------------

plt.figure(figsize=(12,6))
sns.countplot(y="OCCUPATION_TYPE",
              data=final_df,
              order=final_df["OCCUPATION_TYPE"].value_counts().index)
plt.title("Occupation Type")
plt.show()

# ------------------------------------------------------------
# 7. Income Distribution
# ------------------------------------------------------------

plt.figure()
sns.histplot(final_df["AMT_INCOME_TOTAL"], bins=40, kde=True)
plt.title("Annual Income Distribution")
plt.show()

# ------------------------------------------------------------
# 8. Age Distribution
# ------------------------------------------------------------

final_df["AGE"] = (-final_df["DAYS_BIRTH"] // 365)

plt.figure()
sns.histplot(final_df["AGE"], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()

# ------------------------------------------------------------
# 9. Employment Duration
# ------------------------------------------------------------

employment = final_df["DAYS_EMPLOYED"].copy()

employment = employment.replace(365243, np.nan)

employment = (-employment // 365)

plt.figure()
sns.histplot(employment.dropna(), bins=30, kde=True)
plt.title("Employment Experience (Years)")
plt.show()

# ------------------------------------------------------------
# 10. Correlation Heatmap
# ------------------------------------------------------------

numeric_df = final_df.select_dtypes(include=["int64","float64"])

plt.figure(figsize=(12,8))
sns.heatmap(
    numeric_df.corr(),
    cmap="coolwarm",
    annot=False
)

plt.title("Correlation Heatmap")
plt.show()

print("\nEDA Completed Successfully.")
```

```python
# ============================================================
# Step 7 : Feature Engineering
# ============================================================

print("="*80)
print("FEATURE ENGINEERING")
print("="*80)

# Create a copy
df = final_df.copy()

# ------------------------------------------------------------
# Convert Age
# ------------------------------------------------------------

df["AGE"] = (-df["DAYS_BIRTH"] // 365).astype(int)

# ------------------------------------------------------------
# Convert Employment Years
# ------------------------------------------------------------

df["DAYS_EMPLOYED"] = df["DAYS_EMPLOYED"].replace(365243, np.nan)

df["EMPLOYMENT_YEARS"] = (-df["DAYS_EMPLOYED"] / 365).fillna(0)

# ------------------------------------------------------------
# Income Category
# ------------------------------------------------------------

df["INCOME_CATEGORY"] = pd.cut(
    df["AMT_INCOME_TOTAL"],
    bins=[0,100000,200000,500000,1000000,10000000],
    labels=[
        "Low",
        "Lower Middle",
        "Middle",
        "Upper Middle",
        "High"
    ]
)

# ------------------------------------------------------------
# Family Size Category
# ------------------------------------------------------------

df["FAMILY_SIZE"] = pd.cut(
    df["CNT_FAM_MEMBERS"],
    bins=[0,2,4,10],
    labels=[
        "Small",
        "Medium",
        "Large"
    ]
)

# ------------------------------------------------------------
# Remove Unnecessary Columns
# ------------------------------------------------------------

drop_columns = [
    "ID",
    "DAYS_BIRTH",
    "DAYS_EMPLOYED",
    "FLAG_MOBIL"
]

df.drop(columns=drop_columns, inplace=True)

print("\nColumns Removed Successfully.")

print("\nCurrent Shape")
print(df.shape)

print("\nFirst Five Records")
display(df.head())

print("\nFeature Engineering Completed Successfully.")
```

```python
# ============================================================
# Step 8 : Label Encoding (Updated)
# ============================================================

from sklearn.preprocessing import LabelEncoder

print("="*80)
print("LABEL ENCODING")
print("="*80)

encoded_df = df.copy()

label_encoder = LabelEncoder()

# Encode both object and category columns
categorical_columns = encoded_df.select_dtypes(
    include=["object", "category"]
).columns

print("\nCategorical Columns:")
print(list(categorical_columns))

for column in categorical_columns:
    encoded_df[column] = label_encoder.fit_transform(
        encoded_df[column].astype(str)
    )

print("\nEncoding Completed Successfully.")

display(encoded_df.head())

print("\nData Types")

display(encoded_df.dtypes)
```

```python
# ============================================================
# Step 9 : Train-Test Split
# ============================================================

from sklearn.model_selection import train_test_split

print("="*80)
print("TRAIN TEST SPLIT")
print("="*80)

# Features and Target
X = encoded_df.drop("TARGET", axis=1)
y = encoded_df["TARGET"]

print("\nFeature Matrix Shape :", X.shape)
print("Target Vector Shape  :", y.shape)

# Split Dataset
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("\nTraining Set")
print("X_train :", X_train.shape)
print("y_train :", y_train.shape)

print("\nTesting Set")
print("X_test  :", X_test.shape)
print("y_test  :", y_test.shape)

print("\nTrain-Test Split Completed Successfully.")
```

```python
# ============================================================
# Step 10 : Train & Compare Machine Learning Models
# ============================================================

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("="*80)
print("MODEL TRAINING & EVALUATION")
print("="*80)

# ------------------------------------------------------------
# Models
# ------------------------------------------------------------

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(
        n_estimators=200,
        random_state=42
    ),
    "XGBoost": XGBClassifier(
        eval_metric="logloss",
        random_state=42
    )
}

results = []

best_model = None
best_accuracy = 0
best_model_name = ""

# ------------------------------------------------------------
# Training Loop
# ------------------------------------------------------------

for name, model in models.items():

    print("\n" + "="*80)
    print(f"Training : {name}")
    print("="*80)

    # Train
    model.fit(X_train, y_train)

    # Prediction
    y_pred = model.predict(X_test)

    # Metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    results.append([
        name,
        accuracy,
        precision,
        recall,
        f1
    ])

    print(f"Accuracy  : {accuracy:.4f}")
    print(f"Precision : {precision:.4f}")
    print(f"Recall    : {recall:.4f}")
    print(f"F1 Score  : {f1:.4f}")

    print("\nClassification Report")
    print(classification_report(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(5,4))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues"
    )

    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    # Save Best Model
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = model
        best_model_name = name

# ------------------------------------------------------------
# Model Comparison
# ------------------------------------------------------------

results_df = pd.DataFrame(
    results,
    columns=[
        "Model",
        "Accuracy",
        "Precision",
        "Recall",
        "F1 Score"
    ]
)

results_df = results_df.sort_values(
    by="Accuracy",
    ascending=False
)

print("\n")
print("="*80)
print("MODEL COMPARISON")
print("="*80)

display(results_df)

print("\nBest Model :", best_model_name)
print("Best Accuracy :", round(best_accuracy*100,2), "%")
```
